```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Define Tickers and Timeframe
tickers = ['GC=F', 'SI=F'] # Gold and Silver Futures
start_date = '2025-01-01'
end_date = '2026-01-21'

# 2. Download Data
data = yf.download(tickers, start=start_date, end=end_date)['Close']

# 3. Normalize Data (Base 100) to see Percentage Growth
normalized_data = (data / data.iloc[0]) * 100

# 4. Plotting the Horse Race
plt.figure(figsize=(12, 6))
plt.plot(normalized_data['SI=F'], label='Silver (SI=F)', color='#C0C0C0', linewidth=2)
plt.plot(normalized_data['GC=F'], label='Gold (GC=F)', color='#D4AF37', linewidth=2)

# 5. Highlighting the 2026 Record Highs
silver_high = data['SI=F'].max()
gold_high = data['GC=F'].max()

plt.title(f"The 2026 Commodity Horse Race: Silver vs. Gold", fontsize=16)
plt.ylabel("Normalized Price (Start = 100)", fontsize=12)
plt.xlabel("Date", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Add Annotation for the Current Regime
plt.annotate(f'Silver Peak: ${silver_high:.2f}', 
             xy=(normalized_data.index[-1], normalized_data['SI=F'].iloc[-1]),
             xytext=(-150, 20), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='black'))

plt.show()

# Print the Raw Highs for your Records
print(f"2026 Gold High: ${gold_high:,.2f}")
print(f"2026 Silver High: ${silver_high:,.2f}")
```

```python
import yfinance as yf
import pandas as pd

# 1. Define Tickers
tickers = ['GC=F', 'SI=F']

# 2. Download Data
# auto_adjust=True (default) means 'Close' IS the adjusted price now
df = yf.download(tickers, start='2024-12-31', end='2026-01-22')

# 3. Handle MultiIndex Columns
# Structure is now df['Price_Type']['Ticker']
close_prices = df['Close']
high_prices = df['High']

# 4. Extract Key Price Points (2025 Start, 2025 End, 2026 Peak)
si_start_25 = close_prices['SI=F'].loc['2025-01-02'] # First trading day
si_end_25 = close_prices['SI=F'].loc['2025-12-31']   # Last day 2025
si_peak_26 = high_prices['SI=F'].max()               # All-time record

gc_start_25 = close_prices['GC=F'].loc['2025-01-02']
gc_end_25 = close_prices['GC=F'].loc['2025-12-31']

# 5. Calculate Return Percentages
# Surge 2025
silver_2025_return = ((si_end_25 - si_start_25) / si_start_25) * 100
# Surge Trailing (Total Move)
silver_ttm_return = ((si_peak_26 - si_start_25) / si_start_25) * 100
# Gold 2025
gold_2025_return = ((gc_end_25 - gc_start_25) / gc_start_25) * 100

print(f"--- VERIFIED 2026 DATA ---")
print(f"Silver 2025 Surge: {silver_2025_return:.1f}%")
print(f"Silver Trailing Surge: {silver_ttm_return:.1f}%")
print(f"Gold 2025 Surge: {gold_2025_return:.1f}%")
```
